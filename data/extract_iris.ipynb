### Import

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm
import numpy as np
```

## 정부 사업

```python
ancms = []
for idx in tqdm(range(2000,6000)):
    url = f"https://www.iris.go.kr/contents/retrieveBsnsAncmView.do?ancmId=00{idx}"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    ancm = soup.find('ul',class_="list_dot")  
    ancm = str(ancm)
    ancm = ancm.split("\n")
    ancm.append(url)
    ancms.append(ancm)
```

```python
irisDb = pd.DataFrame(columns=["department", "institution", "ancmNo", "title", "ancDate", "period", "link"])
info = []
period = []
link = []
for ancm in tqdm(ancms):
    if len(ancm[1]) == 66:
        continue
    else :
        tmp = []
        for idx in range(1,6,1):
            strIdx = ancm[idx].find("<span>") + 6
            endIdx = ancm[idx].find("</span>")
            tmp.append(ancm[idx][strIdx:endIdx])
        link.append(ancm[10])
        info.append(tmp)
        period.append(ancm[7][ancm[7].rfind("  ")+2:ancm[7].find("</span>")])
```

```python
info = np.array(info)
```

```python
department = info[:,0]
institution = info[:,1]
ancmNo = info[:,2]
title = info[:,3]
ancDate = info[:,4]
```

```python
addTuples = pd.DataFrame({"department": department, "institution" : institution, "ancmNo" : ancmNo, "title" : title, "ancDate" : ancDate, "period" : period, "link" : link})
irisDb = pd.concat([irisDb,addTuples], ignore_index=True)
irisDb
```

```python
irisDb.to_csv("irisBsnAncm.csv", index=False)
```
